# Lists

## Recap

```python
# Lists
Students = ["Eric", "john", "LeXy", "Briam", "Logo", "pipi", "elviS"]
for CurrentStudent in Students:
    print(CurrentStudent.title())
print(Students[-2].title())
```

## Slicing

```python
#ListName[start:stop]
Countries = ["USA", "Cananda", "Mexico", "England", "Jamaica", "Japan", "Kenya", "Tanzania", " Uganda"]
print(Countries [1:4])
print(Countries [3:])
```

# Adding Elements to a List

```python
Countries.append("India")
Countries
```

```python
Countries.insert(0, "Turkey")
Countries
```

## Joining Lists

```python
Countries2 = ["China", "SA"]
Countries + Countries2 
```

## Nested Lists

```python
NestedCountries = [Countries, Countries2]
print(NestedCountries)
```
